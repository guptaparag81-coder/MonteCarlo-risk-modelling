```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
initial_investment = 100000

expected_return = 0.08

volatility = 0.20

time_horizon = 252

num_simulations = 1000
```

```python
simulation_results = np.zeros((time_horizon, num_simulations))

for i in range(num_simulations):

    daily_returns = np.random.normal(
        expected_return/252,
        volatility/np.sqrt(252),
        time_horizon
    )

    price_path = initial_investment * np.cumprod(1 + daily_returns)

    simulation_results[:, i] = price_path
```

```python
plt.figure(figsize=(10,6))

plt.plot(simulation_results)

plt.title("Monte Carlo Portfolio Simulations")

plt.xlabel("Trading Days")

plt.ylabel("Portfolio Value")

plt.grid(True)

plt.show()
```

```python
final_values = simulation_results[-1]
```

```python
var_95 = np.percentile(
    final_values,
    5
)

loss_var = initial_investment - var_95

print("95% Value at Risk (VaR): £",
      round(loss_var,2))
```

```python
plt.figure(figsize=(8,5))

plt.hist(final_values, bins=30)

plt.axvline(
    var_95,
    linestyle='--'
)

plt.title("Distribution of Portfolio Values")

plt.xlabel("Portfolio Value")

plt.ylabel("Frequency")

plt.show()
```
